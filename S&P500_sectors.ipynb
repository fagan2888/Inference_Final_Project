```python
import os
root = os.getcwd() + '/' + 'Inference/'
print(root)
```

```python
import csv
str_date = 20061231
end_date = 20071231
firm_list = {}
with open(root + 'sp500.csv', 'r', newline='') as csvfile:
    firms = csv.reader(csvfile)
    for f in firms:
        if f[0] != 'gvkey':
            if (int(f[3]) <= str_date):
                if (not f[4]):
                    firm_list[f[0]] = f[11]
                elif (int(f[4]) >= end_date):
                    firm_list[f[0]] = f[11]
```

```python
import csv
gics_dict = {}
with open(root + 'gics.csv', 'r', newline='') as csvfile:
    firms = csv.reader(csvfile)
    for f in firms:
        if f[1][0:4] == '2007':
            print(f[0])
            gics_dict[f[0]] = f[15]
```
