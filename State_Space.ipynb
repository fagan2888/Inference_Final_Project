```python
import numpy as np
import statsmodels.api as sm
import pandas as pd
import datetime
```

```python
week_list = []
pd_week =[]
dt = datetime.datetime(2001, 10, 14)
while dt <= datetime.datetime(2007, 6, 24):
    pd_week.append(dt.strftime('%Y-%m-%d'))
    week_list.append(dt)
    dt = dt + datetime.timedelta(7)
    
```

```python
import xlrd
import os
root = os.getcwd() + '/' + 'Inference/'
workbook = xlrd.open_workbook(root + 'S&P500_Sectors.xlsx')
xl_sheet = workbook.sheet_by_index(0)

it_idx = 0
data = np.empty([len(week_list), 11])
info_tech = []
for row in range(5, xl_sheet.nrows):
    dt = datetime.datetime.strptime(xl_sheet.row_values(row)[0], "%d-%b-%Y")
    if it_idx < len(week_list):
        if (dt - week_list[it_idx]) == datetime.timedelta(1):
            temp = xl_sheet.row_values(row)[8]
            temp_row = row
            while temp == 42:
                temp_row = temp_row + 1
                temp = xl_sheet.row_values(temp_row)[8]
            data[it_idx, :] = xl_sheet.row_values(temp_row)[1:]
            it_idx = it_idx + 1
            
```

```python
df = pd.DataFrame(data, index = pd_week)
titles = ['Energy', 'Material', 'Industrials', 'Consumer_Dis', 'Consumer_Sta', 'Health_Care', 
              'Financials', 'IT', 'Telcom', 'Utilities', 'Real_Estate']
df.columns = titles
for tl in titles:
    df['dln_'+tl] = np.log(df[tl]).diff()
```

```python
endog = df.ix['2001-10-21':, 'dln_Energy':'dln_Real_Estate']
mod = sm.tsa.DynamicFactor(endog, k_factor = 1, factor_order = 2, error_order = 1)
```
